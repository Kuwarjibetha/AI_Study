```python
L=['stone','paper','scissor']
chance=int(input('enter chance '))
user_score=0
comp_score=0
while chance!=0:
    comp=np.random.choice(L)
    user=input('enter ')
    if comp==user:
        print('Tie')
    elif comp=='stone':
        if user=='paper':
            print('user wins')
            user_score+=1
        else:
            print('comp wins')
            comp_score+=1

    elif comp=='paer':
        if user=='scissor':
            print('user wins')
            user_score+=1
        else:
            print('comp wins')
            comp_score+=1

    elif comp=='scissor':
        if user=='stone':
            print('user wins')
            user_score+=1
        else:
            print('comp wins')
            comp_score+=1
    chance=chance-1

if comp_score>user_score:
    print('comp wins',comp_score)
elif comp_score==user_score:
    print('Tie')
else:
    print('user wins',user_score)


```
