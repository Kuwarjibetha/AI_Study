```python
# syntax 
# (lambda parameter: expression)(arguments)
```

```python
res=(lambda a,b: a+b)(4,5)
print(res)


res1=lambda x,y:x+y
print(res1(10,3))
```

```python
c=(lambda a,b: a+b)(int(input('enter a ')),int(input('enter b ')))
print(c)
                    

z=(lambda i,j: i-j)
print(z(int(input('enter no.')),int(input('enter no.'))))
```

```python
# lambda para: ans if con else ans
f=(lambda a:'even' if a%2==0 else 'odd')(int(input('enter the value :')))
print(f)
                                         
```

```python
x=(lambda y:'Pallindrome' if y==y[::-1] else 'Not pallindrone')(input('enter strib '))
print(x)
```

```python
s=(lambda a: a.lower() if a.isupper() else a.upper())(input('enter strib '))
print(s)
```

```python
n=int(input('enter n '))
for i in range(n):
    a=int(input('enter a '))
    r=(lambda a:'even' if a%2==0 else 'odd')(a)
    print(a,'is',r)
    
```

```python
# zip- when i want to iterate more than one iterable simultaenously we use zip
```

```python
roll=[1,2,3,4]
name=['a','b','c','d']

for i,j in zip(roll,name):
    print(i,'....',j)
```

```python
roll=[1,2,3,4]
name=['a']

for i,j in zip(roll,name):
    print(i,'....',j)
```

```python
roll=[1,2]
name=['a','b','c','d']
d={}
for i,j in zip(roll,name):
    d[i]=j
print(d)
```
