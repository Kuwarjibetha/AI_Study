```python
# Q1. Write a program to print the following using while loop
# a. First 10 Even numbers
c=0
ev= 2
print("First 10 Even numbers :")

while c<10:
  c=c+1
  print(ev,end=" ")
  ev+=2
print("\n")
# b. First 10 Odd numbers
i =0
od = 1
print("First 10 odd number :")

while i<10:
   i+=1
   print(od , end=" ")
   od+=2
print("\n")

# c. First 10 Natural numbers
nu = 0
print("First 10 Natural numbers :")
while nu<=10:
  nu+=1
  print(nu,end=" ")
print("\n")

# d. First 10 Whole numbers
nu = 0
print("First 10 Whole numbers :")
while nu<10:
  print(nu,end=" ")
  nu+=1
print("\n")

```

```python
# 6. Write a program to enter the numbers till the user enter ZERO and at the end it should display the count and sum of positive  , negative numbers  entered.

pos_count = 0
neg_count = 0
pos_sum = 0
neg_sum = 0

while True:
    user = int(input("enter a number(type 'stop' to fhinish): "))
    if user == 0:
     break
    else:
      if user >0:
        pos_count+=1
        pos_sum +=user
      else:
        neg_count+=1
        neg_sum+= user

print(f"Total positive count {pos_count} and sum: {pos_sum}")
print(f"Total negative count {neg_count} and sum: {neg_sum}")
```

```python
# Q7. Write a program to display all the numbers which are divisible by 13 but not by 3 between 100 and 500.(exclusive both numbers)

i = 100

while i<=500:
  if i%13==0 and i%3!=0:
    print(i,end=" ")
  i+=1
```

```python

# Q10. Write a program to display the number names of the digits of a number entered by user,
# for example if the number is 231
# then output should b e
# Two
# Three
# One

user = input("enter a number: ")

for i in user:
  if i == "0":
    print("Zero")
  elif i == "1":
    print("One")
  elif i == "2":
    print("Two")
  elif i == "3":
    print("Three")
  elif i == "4":
    print("Four")
  elif i == "5":
    print("Five")
  elif i == "6":
    print("Six")
  elif i == "7":
    print("Seven")
  elif i == "8":
    print("Eight")
  else:
    print("Nine")
```
