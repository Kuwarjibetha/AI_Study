```python
def func(L):
  even_sum=0
  odd_sum=0
  div_sum=0

  for i in L:
    if i%2==0:
      even_sum=even_sum+i
    if i%2!=0:
      odd_sum=odd_sum+i
    if i%3==0:
      div_sum=div_sum+i
  return even_sum,odd_sum,div_sum
```

```python
L=[11,14,21,23,56,78,29,28]

func(L)
```

```python
def f1(f2,L):
  res=0
  for i in L:
    if f2(i):
      res=res+i

  return res
```

```python
L=[11,14,21,23,56,78,29,28]
x = lambda x:x%2==0
y = lambda x:x%2!=0
z = lambda x:x%3==0

print(f1(x,L))
print(f1(y,L))
print(f1(z,L))
```

```python
from functools import reduce
```

```python
L=[1,2,3,4,5]
res=lambda x,y:x+y
reduce(res,L)
```

```python
L1=[12,34,56,11,21,58]
res1=lambda x,y:x if x>y else y
reduce(res1,L1)
```
