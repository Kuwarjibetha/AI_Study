# 2. creating Table

```python
qry='''create table college (
clg_id int, 
name varchar(255), 
estd_year int)'''
```

```python
cur.execute(qry)
print('Database university has a table by name colleg')
```

```python
qry='''insert into college(clg_id, name, estd_year)
values(101,'DU',2000),
(102,'IIT',1980),
(103,'GLA',2017),
(104,'Amity',2010),
(105,'sharda',2015)
'''

cur.execute(qry)

```
