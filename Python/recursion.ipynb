```python
def mul(a,b):
  if b==1:
    return a
  else:
    return a+mul(a,b-1)

mul(2,3)
```

```python
def fact(n):
  if n==1:
    return 1
  else:
    return n* fact(n-1)

fact(4)
```

```python
def plain(text):
  if len(text)==1:
    print('pallindrome')
  else:
    if text[0]==text[-1]:
      plain(text[1:-1])
    else:
      print('Not a palin')
```

```python
plain('mam')
```
