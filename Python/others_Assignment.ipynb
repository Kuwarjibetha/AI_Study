```python
# 8. Write a python program to calculate the hypotenuse of a right angled triangle.
```

```python
a=float(input("Enter the length of the first side (a): "))
b=float(input("Enter the length of the second side (b): "))

c_squared=a*a+b*b
hypotenuse=0.5*(c_squared + c_squared/2)
hypotenuse=0.5*(hypotenuse + c_squared/hypotenuse)

print(f"The hypotenuse of the triangle is: {hypotenuse}")
```
