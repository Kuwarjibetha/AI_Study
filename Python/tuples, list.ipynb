# list Comprehension

```python
# when i want my exprected output in the form of list we use list compr
# we use list comprehesnion for a shorter syntax of code
```

```python
L=[]
for i in range(1,10):
    L.append(i)
print(L)
```

```python
# sytnax [ans for loop]
l1=[i for i in range(1,10)]
print(l1)
```

```python
L=[]
for i in range(1,10):
    if i>6:
        L.append(i)
print(L)
```

```python
# syntx [ans for loop if con]
l2=[i for i in range(1,10) if i>6]
print(l2)
```

```python
L=[]
for i in range(1,10):
    if i%2==0:
        L.append('even')
    else:
        L.append('odd')
print(L)
```

```python
# [ans if con else ans for loop]
l3=[i if i%2==0 else 'odd' for i in range(1,10)]
print(l3)
```
