```python
d={}
for i in range(1,10):
    # d[i]=i+1
    d.update({i:i+1})
print(d)
```

```python
# {key:value for loop }
d1={i:i+1 for i in range(1,10)}
print(d1)
```

```python
d={}
for i in range(1,10):
    if i>6:
        d[i]=i-1
print(d)
```

```python
# {key:value for loop if con}

d2={i:i-1 for i in range(1,10) if i>6}
print(d2)
```

```python
d={}
for i in range(1,10):
    if i%2==0:
        d[i]='even'
    else:
        d[i]='odd'
print(d)
```

```python
# {key: value if con else value for loop}
d3={i:'even' if i%2==0 else 'odd' for i in range(1,10)}
print(d3)
```
