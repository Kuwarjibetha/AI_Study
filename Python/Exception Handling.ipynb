```python
a=int(input('enter a '))
b=int(input('enter b '))

try:
    c=a/b
    print(c)

except:
    print('Denoimantor cant be zero')
    
```

```python
# try block always runs whether there is an exception or not
# excpet block runs only when there is an excpetion
```

```python

try:
    a=int(input('enter a '))
    b=int(input('enter b '))

    c=a/b
    print(c)

except ZeroDivisionError:
    print('Denoimantor cant be zero')

except ValueError:
    print('Only pass the integer')
    
```

```python
try:
    L=[1,2,3]
    a=int(input('enter a '))
    b=int(input('enter b '))

    c=a/b
    print(c)
    print(L[b])


except ZeroDivisionError:
    print('Denoimantor cant be zero')

except ValueError:
    print('Only pass the integer')

except IndexError:
    print('index out of range')


    

    
```
