# Seaborn

```python
# seaborn - it is used to perform the statistical analysis
        # we can plot the mutiple columns in the graph
        # we can aslo perform the aggreagion functions in the graph
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
# Numerical data 
#1. relplot
#2. line plot
#3. scatterplot
```

```python
# relplot - 
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',style='size',col='time',row='smoker',data=df);
```

```python
df['size'].unique()
```

```python
sns.lineplot(x='total_bill',y='tip',data=df);
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df);
```

```python
# Categorical data
# 1. Catplot
# 2. Swarmplot
# 3. barplot
# 4. histplot
#    countplot
# 5. boxplot
```

```python
df
```

```python
sns.catplot(x='day',y='total_bill',hue='day',data=df); # striplot
```

```python
sns.catplot(x='day',y='total_bill',hue='day',kind='swarm',data=df); # swarm plot
```

```python
df.groupby(['day'])['total_bill'].agg(['sum','mean','min'])
```

```python
sns.barplot(x='day',y='total_bill',ci=None,estimator='min',data=df);
```

```python
sns.catplot(x='day',y='total_bill',hue='day',kind='bar',ci=None,data=df); # swarm plot
```

```python
df['size'].value_counts()
```

```python
sns.histplot(df['size']);
```

```python
sns.countplot(x='day',hue='day',data=df)
```

```python
sns.catplot(x='day',hue='day',kind='count',data=df)
```

```python
sns.boxplot(x='tip',data=df);
```

```python
sns.boxplot(x='day',y='tip',hue='sex',data=df);
```

```python
sns.catplot(x='day',y='tip',hue='sex',kind='a',data=df);
```
