# Boxplot

```python
# it is used to detect the outliers in the data
```

```python
plt.boxplot(df1['tip']);
```

```python
# q0=0%
# q1=25%
# q2=50%
# q3=75%
# q4=100%


# upper_limit = q3+1.5*(IQR)
# lower_limit = q1-1.5*(IQR)

# Inter Quartile Range= q3-q1
```

```python
df1['tip'].describe()
```

```python
q1=df1['tip'].quantile(0.25)
q1
```

```python
q3=np.percentile(df1['tip'],75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
ul=q3+1.5*(IQR)
ul
```

```python
ll=q1-1.5*(IQR)
ll
```

```python
df1
```

```python
df1[df1['tip']<ul]
```

```python
df1[df1['tip']<ll]
```

```python
#1. Trimming
#2. Capping/ winsorization
```

```python
df1['new_tip']=np.where(df1['tip']>ul,ul, 
                        np.where(df1['tip']<ll,ll,
                                    df1['tip']))
df1
```

```python
plt.boxplot(df1['new_tip']);
```
